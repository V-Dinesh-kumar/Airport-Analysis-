```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
df=pd.read_excel('Airports.xlsx')
print(df)
```

```python
# countries by total passenger
ap=pd.DataFrame(df.groupby('Country').sum()['Passengers'])
print(ap)
```

```python
# most crowded airport bar chart
ap.plot(kind="bar",figsize=(18,7),color = ("grey"))
plt.xticks(rotation=90)
plt.xlabel("Country",fontsize=20)
plt.ylabel("Passenger count in billions",fontsize=20)
plt.title("Most crowded Airport countries",fontsize=20);
```

```python
# Displaying sum of passenger list by year(table)
Yp=pd.pivot_table(df,index=('Year'),values=['Passengers'])
print(Yp)
```

```python
# Passenger count by years bar chart
Yp.plot(kind="bar",figsize=(18,10),color = ("Blue"))
plt.xticks(rotation=0,fontsize=20)
plt.yticks(rotation=0,fontsize=20)
plt.xlabel("Year",fontsize=25)
plt.ylabel("Passenger count in billions",fontsize=20)
plt.title("Passenger count by years",fontsize=20)
```

```python
# most count of passengers list by Year
mp=pd.pivot_table(df,index=['Year'],values=['Passengers'], aggfunc=('sum')).sort_values('Passengers',ascending=False)
mp.head(2)
```

```python
# most count of passengers list by country
mc=pd.pivot_table(df,index=['Country'],values=['Passengers'], aggfunc=('sum')).sort_values('Passengers',ascending=False)
mc.head(3)
```

```python
# most count of passengers list by location
ml=pd.pivot_table(df,index=['Location'],values=['Passengers'], aggfunc=('sum')).sort_values('Passengers',ascending=False)
ml.head(3)
```

```python
# Least count of passengers list by country
mc.tail(3)
```
